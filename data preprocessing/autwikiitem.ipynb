```python
import numpy as np
import pandas as pd
import os
import csv
```

```python
os.chdir("path to item_marcrealaut.csv")
```

```python
df = pd.read_csv('item_marcrelaut.csv',delimiter=' ,,, ',header=None)
df.head()
```

generate a new column and leave only the gnd number to align the result from Wikidata query.

```python
df[2]=df[1]
```

```python
df[2] = df[2].map(lambda x: str(x)[22:])
```

```python
df[2] = df[2].map(lambda x: str(x)[:-1])
```

```python
#rename the columns
df.columns = ['item', 'gnduri','gnd']
```

```python
df.head()
```

```python
#read 'gndquery.csv'
df2=pd.read_csv('gndquery.csv')
```

```python
df2
```

```python
#merge on 'gnd' and 'GND__' to find out all the items whose authors exist in Wikidata
autwiki=df.merge(df2,left_on='gnd',right_on='GND__',how="inner")
```

```python
autwiki['item_x'].value_counts()
```

```python
df.to_csv('item_gnduri_gnb.csv',header=None,index=0)
```

```python
#save the item column in a new dataframe
autwikiitem=autwiki['item_x'].to_frame()
```
